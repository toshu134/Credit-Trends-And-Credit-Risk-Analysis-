```python
import pandas as pd
import matplotlib.pyplot as plt

#factors affecting credit score
data = pd.read_csv('your_data.csv')
correlation_matrix = data.corr()
plt.figure(figsize=(12, 12))

# Create colormap for the heatmap
cmap = 'coolwarm'  # You can choose other colormaps like 'RdGy', 'RdYlGn', 'Blues'

# Generate heatmap using matplotlib
plt.imshow(correlation_matrix, cmap=cmap)

# Add colorbar to the plot
plt.colorbar()

# Add labels for each cell (optional)
for i in range(len(correlation_matrix.columns)):
    for j in range(len(correlation_matrix.columns)):
        plt.text(i, j, correlation_matrix.iloc[i, j], ha='center', va='center')

# Add labels for rows and columns
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)

# Set plot title (optional)
plt.title('Correlation Matrix of Loan Data')

# Show the plot
plt.show()



#home buying seasonlity over the years
month_data = data['Month']

# Assuming 'Number of Home Purchases' is the column with purchase data (adjust column names if needed)
purchase_data = data['Number of Home Purchases']

# Create the line plot
plt.figure(figsize=(10, 6))  # Adjust figure size as desired
plt.plot(month_data, purchase_data)

# Set labels for x and y axes (adjust based on your data)
plt.xlabel('Month')
plt.ylabel('Number of Home Purchases')

# Set title for the graph (optional)
plt.title('Home Buying Seasonality Over the Years')

# Rotate x-axis labels for better readability if there are many data points
plt.xticks(rotation=45)

# Show the plot
plt.grid(True)  # Add grid lines for better readability (optional)
plt.show()

```
